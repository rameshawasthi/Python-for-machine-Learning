```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url) #soup variable
soup
```

```python
soup = soup.text
soup
```

```python
soup = BeautifulSoup(soup, "lxml")
soup
```

```python
table = soup.table
table   #variable
```

```python
tags = table.find_all('tr')
tags # variable
```

```python
data = []
for tag in tags:
    x = tag.text
    x = x.replace('\n','')   # replace \n ;
    data.append(x)
print(data)
```

```python
data = []
for tag in tags:
    x = tag.text
    x = x.split('\n')   
    data.append(x)
print(data)
```

```python
final_data = []
for j in data:
    if j[0]!='':
        final_data.append(j)
print(final_data)
```

```python
import csv
file = open('covid_data.csv', 'w')
x = csv.writer(file)  # x is object
for i in final_data:
    x.writerow(i)
file.close()
```
